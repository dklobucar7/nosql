```python
data = pd.read_csv("hungary_chickenpox.csv")

# Funkcija za kalklaciju IV i WoE vrijednosti
def calculate(dataset, feature):
    
    # Definiramo dobre i lose vrijednost, dobre one koje ulazu u 73. percentil a lose one izvan njega
    Q1 = dataset[feature].quantile(0.73)

    dataset['target'] = np.where(dataset[feature] <= Q1, 0, np.where(dataset[feature] >= Q1, 1, -1))

    results = []
    total_good = dataset[feature][dataset['target'] == 0].count()
    total_bad = dataset[feature][dataset['target'] == 1].count()

    # Izracun WoE i IV
    prop_good = total_good / (total_good + total_bad)
    prop_bad = total_bad / (total_bad + total_good)
    WoE = np.log(prop_good / prop_bad)
    IV = (prop_good - prop_bad) * WoE
    results.append((IV, WoE))
    
    return [res[0] for res in results][0], [res[1] for res in results][0]

# Petlja za izracun po svim kolonama
results_quartiles = {}
for column in data.columns:
    if column != 'Date':
        iv, woe = calculate(data, column)
        results_quartiles[column] = {'IV': iv, 'WoE': woe}

final_results_df = pd.DataFrame(results_quartiles).T
final_results_df.sort_values(by='IV', ascending=False)

# Sve varijable imaju WoE vrijednost vecu od 0, sto ukazuje na jaku prediktivnu moc

# Vecina varijabla ima vrijednost IV 0.45 - 0.5 sto ukazuje na jaku prediktivnu moc,
# Za varijable Baranya, Bekes i Somogy vrijednosti su > 0.5 sto moze znaciti sumnjivo ponaanje,
# no s obizrom na dataset visoke IV vrijednosti su ocekivane
```

```python
# Kreiranje grafa za top 5 varijabli po WoE
WoE_values = final_results_df.sort_values(by='WoE', ascending=False).head(5)['WoE'].values
variables = final_results_df.sort_values(by='WoE', ascending=False).head(5).index.values

plt.figure(figsize=(10, 8))
plt.bar(variables, WoE_values, color='skyblue')
plt.ylabel('WoE')
plt.title('WoE values for each variable')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()

# Sve varijable imaju WoE vrijednost vecu od 0, sto ukazuje na jaku prediktivnu moc
```

```python
# Kategoriziramo varijable u razrede po WoE vrijednosti:
#     < - 1       -> Niska prediktivna moc
#     > -1 & < 1  -> Srednja prediktivna moc
#     > 1         -> Visoka prediktivna moc

woe_ranges = {
    'Low': (-np.inf, -1),
    'Medium': (-1, 1),
    'High': (1, np.inf)
}

grouped_regions_df = pd.DataFrame(columns=['Region', 'WoE Group', 'IV', 'WoE'])

for region, row in final_results_df.iterrows():
    woe_value = row['WoE']
    iv_value = row['IV']
    woe_group = None
    for group, (low, high) in woe_ranges.items():
        if low <= woe_value <= high:
            woe_group = group
            break
    new_row = pd.DataFrame({'Region': [region], 'WoE Group': [woe_group], 'IV': [iv_value], 'WoE': [woe_value]})
    grouped_regions_df = pd.concat([grouped_regions_df, new_row], ignore_index=True)

grouped_regions_df.sort_values(by='IV', ascending=False)

# Iz prilozenog vidimo da sve varijable spadaju ili u grupu Visoke ili Srednje prediktivne moci,
# sto znaci da ce nam svaka varijabla dobro doci u budicim prediktivnim modelima
```

```python
df = pd.read_csv("hungary_chickenpox.csv")
df['Total Cases'] = df.drop('Date', axis=1).sum(axis=1)

percentiles = df.describe(percentiles=[0.25, 0.75]).loc[['25%', '75%']].to_dict()

def categorize(value, col_name):
    if value <= percentiles[col_name]['25%']:
        return 'low'
    elif value <= percentiles[col_name]['75%']:
        return 'medium'
    else:
        return 'high'

for col in df.columns[1:]:
    df[col] = df[col].apply(lambda x: categorize(x, col))

# Buduci da je dataset pun brojeva koji oznacavaju broj slucajeva vodenih kozica kroz regije u Madarskoj
# kao takav nije pogodan za Bayesovu mrezu pa transformiramo podatke da ih mozemo dalje koristiti, za transformaciju
# odlucio sam se za svaki redak vrijednost oznaciti kao low, high i medium ovisno o percentilima u kojima se vrijednost nalazi,
# <= 25 % -> Low
# <= 75 % -> Medium
# >  75 % -> High

df.drop(['Date'], axis=1).to_csv("data_for_bayes.csv", index=False)
```
