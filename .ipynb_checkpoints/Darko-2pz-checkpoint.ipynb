### 1. Kontrola kvalitete podataka

#### Uvoz podataka

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plot
%matplotlib inline
novosti = pd.read_csv("News_Final_500.csv")
novosti.head()
```

#### Prepoznavanje nedostajućih vrijednosti 

```python
novosti.drop(novosti.columns[[0,1]], axis=1, inplace=True) #brisanje prva dva nepotrebna stupca
nedostajuce_vrijednosti = novosti[novosti.isna().any(axis=1)] #axis=1 - operacije se obavaljaju horizontalno; 0 vertikalno
nedostajuce_vrijednosti.head()
```

#### Rekodiranje nedostajućih vrijednosti

```python
novosti.fillna("missing", inplace=True) #inplace - promjene se odmah reflektiraju na orginalni dataset
```

```python
novosti[novosti.isna().any(axis=1)] #provjera ima li još uvijek nedostajućih vrijednosti
```

#### Prepoznavanje unifromnih ili nelogičnih distribucija

```python
#primjer koristeći describe
novosti.describe()
```

Među atributima "Facebook", "GooglePlus" i "LinkedIn" vidljivo je da ja velika standardna devijacija što upućuje na nelogičnu distribuciju dok atributi "SentimentTitle" i "SentimentHeadline" pokazuju uniformnu distribuciju - mala standardna devijacija. Također je vidljivo da u sva tri atributa, "Facebook", "GooglePlus" i "LinkeIn", imaju ekstremne vrijednosti tj. koje su vece za 3 standardne devijacije (max vrijednosti) od srednje vrijednosti. Moglo bi se reći da imamo ekstremno popularne vijesti(velik broj dijeljnja, lajkova, komentara itd.). Vrijednosti u prvom kvartilu za "Facebook", "GooglePlus" i "LinkedIn" su 0 što znači da 25% podataka u tim stupcima ima vrijednost nula ili manje, a 75% vrijednosti veću vrijednost od nule.

```python
#dohvat i ispis 3 najekstremije vrijednosti za atribute "Facebook", "GooglePlus" i "LinkedIn"
for drustvena_mreza in ["Facebook", "GooglePlus", "LinkedIn"]:
    maks_vrijednost = novosti[drustvena_mreza].max()
    redovi_s_maks_vrijednoscu = novosti[novosti[drustvena_mreza] == maks_vrijednost]
    naslovi = redovi_s_maks_vrijednoscu["Title"].tolist()
    print(f"Najveća vrijednost za {drustvena_mreza}: {maks_vrijednost}")
    print("Naslov s najvećom vrijednošću:")
    for naslov in naslovi:
        print(f"- {naslov}")
    print("\n")
```

Vizualno, distribuciju možemo prikazati histogramima. Vidljivo je da npr. atribut "LinkedIn" ima nelogičnu distribuciju tj. ima ektremne vrijednosti tj. vrijednosti koje 3 i više stadardnih devijacija odstupaju od srednje vrijednosti. Isto vrijedi i za atribute "GooglePlus" i "Facebook".

```python
graf = novosti[novosti["Topic"] == "economy"]["LinkedIn"].plot(kind="hist",bins=20, color='purple')
graf.set_xlabel('LinkedIn')
graf.set_ylabel('Broj vijesti')
graf.set_title('Histogram za LinkedIn (Topic: economy)')
```

Atribut "SentimentTitle" ima uniformnu distribuciju, a kao što je vidljivo u tablici gore, standardna devijacija je mala. (.describe()). Isto vrijedi i za atribut "SentimentHeadline".

```python
graf1 = novosti[novosti["Topic"] == "economy"]["SentimentTitle"].plot(kind="hist",bins=20, color='purple')
graf1.set_xlabel('SentimentTitle')
graf1.set_ylabel('Broj vijesti')
graf1.set_title('Histogram za SentimentTitle (Topic: economy)')
```

#### Određivanje ciljne varijable

Atributi "Facebook", "GooglePlus" i "LinkedIn" predstavljaju konačnu popularnost koju je vijest postigla na toj društvenoj mreži unutar 2 dana od objave. Proizvoljno je određeno da će se vijest smatrati popularnom ako je vrijednost popularnosti na bilo kojoj društvenoj mreži prešla broj 5.

### 2. Statističke analize uzorka podataka

#### Distribucija frekvencije za svaku kategoričku varijablu

```python
kategoricke_varijable = ['Popularnost','Topic','Source','Title','Headline','PublishDate']
for varijabla in kategoricke_varijable:                       
    graf = novosti[varijabla].value_counts().plot(kind='bar', color='skyblue')
    plot.title(f'Distribucija frekvencije za  {varijabla}')
    plot.ylabel('Frekvencija')
    plot.show()
```

Vidljivo je da je broj popularnih vijesti podjednak s onim koji nisu popularne. Broj vijesti vezanih za ekonomiju je najveći dok je broj vijesti vezanih za Palestinu najmanji. Atributi koji u sebi imaju velik broj različitih vrijednosti - gotovo je nemoguće očitati iz histograma pojedinu vrijednost.

#### Izračun srednje vrijednosti, zaobljenosti distribucije, nagnutosti distribucije te Z transformacija za sve kontinuirane varijable

```python
numericke_varijable = ['SentimentTitle','SentimentHeadline','Facebook','GooglePlus','LinkedIn']
for varijabla in numericke_varijable:
    sv = novosti[varijabla].mean()
    zaobljenost_distribucije = novosti[varijabla].kurt()
    nagnutost_distribucije = novosti[varijabla].skew()
    print(f"Srednja vrijednost - {varijabla}: {sv}")
    print(f"Zaobljenost distribucije - {varijabla}: {zaobljenost_distribucije}")
    print(f"Nagnutost distribucije - {varijabla}: {nagnutost_distribucije}")
    print("")
   
#Z transformacija
novosti["st_z"] = (novosti["SentimentTitle"]-novosti['SentimentTitle'].mean())/(novosti["SentimentTitle"].std())
novosti["sh_z"] = (novosti["SentimentHeadline"]-novosti['SentimentHeadline'].mean())/(novosti["SentimentHeadline"].std())
novosti["fb_z"] = (novosti["Facebook"]-novosti['Facebook'].mean())/(novosti["Facebook"].std())
novosti["go_z"] = (novosti["GooglePlus"]-novosti['GooglePlus'].mean())/(novosti["GooglePlus"].std())
novosti["li_z"] = (novosti["LinkedIn"]-novosti['LinkedIn'].mean())/(novosti["LinkedIn"].std())
novosti.head()
```

Vidljivo je da atributi "Facebook", "GooglePlus" i "LinkedIn" imaju visoku distribuciju zaobljenosti što ukazuje na postojanje velikih ekstremnih vrijednosti. Pozitivna nagnutost distribucije za iste atribute da imaju više ekstremnih vrijednosti s desne strane srednje vrijednosti. Kod atributa "SentimentTitle" i "SentimentHeadline" nagnutost distribucije je na desno što govori da imaju više vrijednosti s lijeve strane srednje vrijednosti.

#### Grafikon "Z" obilježja za atribut "Facebook" 

```python
#izvršavanje koda traje malo duže
graf_fb = novosti['fb_z'].plot(kind='bar', color='skyblue')
graf_fb.set_title('Grafikon "Z" obilježja za atribut "Facebook"')
```

Na grafikonu "Z" obilježja se primjećuje postojanje ekstremnih vrijednosti tj onih kojih je vrijednost veća od srednje vrijednosti plus tri standardne devijacije.

```python
#selekcija slogova koji imaju "Z" vrijednost (obilježje) >= 3
print(novosti.loc[novosti['fb_z'] >= 3])
```

#### Izračun korelacije svake varijable s ciljnom varijablom

```python
#Pretvaranje ciljne varijable u numeričku
novosti['Popularnost1'] = 0
novosti.loc[(novosti['Facebook'] > 5) | (novosti['GooglePlus'] > 5) | (novosti['LinkedIn'] > 5), 'Popularnost1'] = 1
#korelacija s ciljnom varijablom
novosti.corr()['Popularnost1']
```

Najveću korelaciju s ciljnom varijablom ima atribut "GooglePlus" dok najmanju korelaciju ima atribut "SentimentTitle". 

#### Izračun korelacije svake varijable izražene kroz Z vrijednost s ciljnom varijablom i  kategoriziranje u razrede <-3 , (-3 do 3),>3 

```python
granice_intervala = [float('-inf'), -3, 3, float('inf')]
mapiranje = {'<-3': 0, '(-3 do 3)': 1, '>3': 2}

z_vrijednosti = ['st_z','sh_z','fb_z','go_z','li_z']
for varijabla in z_vrijednosti:
    novi_stupac = varijabla +' _kategorija'
    novi_stupac1 = varijabla +' _kategorija1'
    novosti[novi_stupac] = pd.cut(novosti[varijabla], bins=granice_intervala, labels=['<-3', '(-3 do 3)', '>3'])
    novosti[novi_stupac1] = novosti[novi_stupac].map(mapiranje)
    korelacija = novosti[novi_stupac1].corr(novosti['Popularnost1'])
    print(f'Korelacija {varijabla} sa ciljnom varijablom je: {korelacija} ' )
```

Najveću korelaciju (izraženu kroz Z vrijednost) s ciljnom varijablom ima atribut "li_z"(Z obilježje od atributa "LinkedIn") dok najmanju korelaciju s ciljnom varijablom ima atribut "st_z" (Z obilježje od atributa "SentimentTitle"). 

## 3. IV, WoE izračuni

#### Izračun IV, WoE za svaku varijablu 

```python
numericke_varijable = ['SentimentTitle','SentimentHeadline','Facebook','GooglePlus','LinkedIn']
FB = None
GO = None
LI = None
for varijabla in numericke_varijable:
    varijabla_kategorija = varijabla + '_kategorija'
    novosti[varijabla_kategorija] = pd.qcut(novosti[varijabla], 10, duplicates='drop')
    a = novosti[varijabla_kategorija]
    b = novosti['Popularnost']
    IV = pd.crosstab(a,b).apply(lambda c: c/c.sum(), axis=0)
    IV['WoE'] = np.log(IV['DA']/IV['NE'])
    IV['IV'] = (IV['DA']-IV['NE'])* np.log(IV['DA']/IV['NE'])
    
    print(IV)
    print('')
    print(f'Suma IV: {IV["IV"].sum()}')
    print('')
    #izdvajanje varijabli s najjačim IV
    if varijabla == 'Facebook':
        FB = IV
    if varijabla == 'GooglePlus':
        GO = IV
    if varijabla == 'LinkedIn':
        LI = IV
```

Vidljivo je da su atributi "SentimentTitle" i "SentimentHeadline" sadrže kategorije kod kojih se WoE "vrti" oko nula pa su manje relevantni za predekciju popularnosti. Atributi "Facebook", "GooglePlus" i "LinkedIn" sadrže kategorije koje utječu na nepopularnost, popularnost, ali i one kategorije koje su irelevantne za predikciju ciljne varijable.  Kod navedena tri "jaka" atributa, suma vrijednosti IV je beskonačna dakle imaju neizmjerno velik utjecaj na predikciju (suma IV = inf). Obratno suma IV za sentiment varijable je vrlo mala i stoga imali vrlo mali utjecaj na ciljnu varijablu.

Cheatsheet:

Kategorije kod kojih su WoE(vrijednost težine dokaza) blizu nuli, su irelevantnije za predikciju popularnosti. Kategorije koje imaju negativnu vrijednost utječu na nepopularnost. Što je vrijednost kategorije (WoE) veća od nula ta kategorija više utječe na popularnost.
Atributi koji imaju višu vrijednost informacije (IV), zanačajniju su za predikciju ciljne varijable.

Utjecaj IV na ciljnu varijablu:
    0 - ne utječu; 
    0,2 - 0,6 - umjereno utječu; 
    > 0,6 značajno utječu;
    > 1 značajno utječu te moraju biti testirani na korelaciju s ciljnom varijablom

#### Izdvajanje varijable sa najjačim IV-om i grafički prikaz WoE za sve najjače varijable

```python
najjace_varijable = ['Facebook','GooglePlus','LinkedIn']
for varijabla in najjace_varijable:
    if varijabla == 'Facebook':
        FB[['WoE']].plot(kind='bar')
        plot.show()
    if varijabla == 'GooglePlus':
        GO[['WoE']].plot(kind='bar')
        plot.show()
    if varijabla == 'LinkedIn':
        LI[['WoE']].plot(kind='bar')
        plot.show()
```

#### Grupiranje u razrede prema kriteriju WoE za najjače varijable

Prema grafikonima/histogramima atribut "Facebook" bi se grupirao u četiri kategorije (-1,2),(2,4),(4,10) i (10,inf).
Atribut "GooglePlus" bi se grupirao u 4 kategorije (-1,0),(0,1),(1,9),(9,inf). Atribut "LinkedIn" bi se grupirao u 3 kategorije (-1,0),(0,8),(8,inf). Navedene grupe će biti korištene u 3. projektnom zadatku. Sentiment atributi "SentimentTitle" i "SentimentHeadline" će biti podjeljeni u dvije kategorije, pozitivnu i negativnu "< 0" i "> 0".
